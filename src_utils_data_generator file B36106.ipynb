```python
# Synthetic arrival profile generator (for future use)
import numpy as np

def generate_arrival_profile(days=30, base_lambda=10, peak_multiplier=2.5):
    """Generate simple time-of-day arrival multipliers for synthetic data."""
    profiles = []
    for _ in range(days):
        day_profile = []
        for hour_block in range(3):
            lam = base_lambda * (1 + (peak_multiplier - 1) * (1 if hour_block == 1 else 0))
            day_profile.append(lam)
        profiles.append(day_profile)
    return np.array(profiles)
```
